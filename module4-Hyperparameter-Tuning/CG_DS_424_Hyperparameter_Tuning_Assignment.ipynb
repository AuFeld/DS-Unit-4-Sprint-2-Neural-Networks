```python
# imports

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import category_encoders as ce

# alter disply output
pd.options.display.max_columns = 999
```

```python
# load data

df = pd.read_csv("/content/WA_Fn-UseC_-Telco-Customer-Churn+(1).csv")
```

```python
# create and drop N/A values

df = df.replace(" ", np.NaN).dropna()
```

```python
df.head()
```

```python
# seperate target variable and drop target variable from df

y = df['Churn'].replace({'Yes':1, 'No':0})
df = df.drop('Churn', axis=1)
```

```python
# split train / test

X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.25, 
                                                    stratify=y, random_state=7)
```

```python
# look at baseline

y_train.value_counts(normalize=True)
```

```python
X_train.describe(include='all')
```

```python
# convert column to a float

X_train['TotalCharges'] = X_train['TotalCharges'].astype('float')
X_test['TotalCharges'] = X_test['TotalCharges'].astype('float')
```

```python
# drop id column

X_train = X_train.drop('customerID', axis=1)
X_test = X_test.drop('customerID', axis=1)
```

```python
# one hot encode categoricals

encoder = ce.OneHotEncoder(use_cat_names=True)
X_train = encoder.fit_transform(X_train)
X_test = encoder.fit_transform(X_test)
```

```python
# normalize numeric data on a scale of 0-1

X_train = X_train / np.amax(X_train, axis=0)
X_test = X_test / np.amax(X_train, axis=0)
```

```python
# check yo data

X_train.head()
```

```python
X_train.shape
```

```python
# fit baseline keras model

from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# pre-specify hyperparamters

inputs = X_train.shape[1]
epochs = 50
batch_size = 100

# create model

model = Sequential()
model.add(Dense(45, activation='relu', input_shape=(inputs,)))
model.add(Dense(45, activation='relu'))
model.add(Dense(45, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# compile model

model.compile(optimizer='nadam', 
              loss='binary_crossentropy', 
              metrics=['acc'])

# fit model

model.fit(X_train, y_train, 
          validation_data=(X_test, y_test), 
          epochs=epochs,
          batch_size=batch_size)
```

```python
# hyperparameter tuning

from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras import optimizers

# pre-specify hyperparameters
inputs = X_train.shape[1]
# epochs = 50
# batch_size = 100

# create Model

def create_model():
    model = Sequential()
    model.add(Dense(45, activation='relu', input_shape=(inputs,)))
    model.add(Dense(45, activation='relu'))
    model.add(Dense(45, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='nadam', 
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])
    return model

# prep for Grid Search

model = KerasClassifier(build_fn=create_model, verbose=0)

# define Grid Search params

param_grid = {'batch_size': [20, 60, 80, 100, 200],
             'epochs': [20]}

grid = GridSearchCV(estimator=model, 
                    param_grid=param_grid,
                   n_jobs=1)
grid_result = grid.fit(X_train, y_train)
```

```python
# report results

print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")

means = grid_result.cv_results_['mean_test_score']

stds = grid_result.cv_results_['std_test_score']

params = grid_result.cv_results_['params']

for mean, stdev, param in zip(means, stds, params):
    print(f"Means: {mean}, Stdev: {stdev} with: {param}")
```

```python
# tune optimizer

from tensorflow.keras import optimizers

# create model

def create_model(learn_rate=0.001):
  model = Sequential()
  model.add(Dense(45, activation='relu', input_shape=(inputs,)))
  model.add(Dense(45, activation='relu'))
  model.add(Dense(45, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(optimizer=optimizers.Adam(lr=learn_rate),
                loss='binary_crossentropy', 
                metrics=['acc'])
  return model

# prepare for grid search

model = KerasClassifier(build_fn=create_model, verbose=0)

# optimize with optimizer

param_grid = {'batch_size': [80], 
              'epochs': [20], 
              'learn_rate': [0.001, 0.01, 0.1, 0.2, 0.3, 0.5]}

grid = GridSearchCV(estimator=model, 
                    param_grid=param_grid, 
                    n_jobs=1)

grid_result = grid.fit(X_train, y_train)
```

```python
# report results

print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print(f"Means: {mean}, Stdev: {stdev} with: {param}")
```

```python
# tune epochs

# create model

def create_model(): 
  model = Sequential()
  model.add(Dense(45, activation='relu', input_shape=(inputs,)))
  model.add(Dense(45, activation='relu'))
  model.add(Dense(45, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(optimizer=optimizers.Adam(lr=0.1),
                loss='binary_crossentropy',
                metrics=['acc'])
  return model

# prepare for grid search

model = KerasClassifier(build_fn=create_model, verbose=0)

# define grid search params

param_grid = {'batch_size': [80], 
              'epochs': [50, 100, 200, 500]}

grid = GridSearchCV(estimator=model, 
                    param_grid=param_grid,
                    n_jobs=1)

grid_result = grid.fit(X_train, y_train)
```

```python
# Report Results

print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print(f"Means: {mean}, Stdev: {stdev} with: {param}")
```
